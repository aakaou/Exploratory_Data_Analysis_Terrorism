```python
# Import libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#link of data
url ='C:/Users/aboub/Downloads/Global Terrorism - START data/globalterrorismdb_0718dist.csv'
```

```python
# Read the dataset
data = pd.read_csv(url, encoding="ISO-8859-1")
```

```python
# Drop unnecessary columns
data.drop(['eventid','approxdate','extended','resolution','country','region','vicinity','specificity','location','summary','crit1','crit2','crit3','doubtterr','multiple','success','suicide','attacktype2','attacktype3','targtype2','targsubtype2','corp2','target2','natlty2','targtype3','targsubtype3','corp3','target3','natlty3','gsubname','gname2','gsubname2','gname3','gsubname3','guncertain1','guncertain2','guncertain3','claimmode2','claimmode3','compclaim','weaptype2','weaptype3','weaptype4','weapsubtype2','weapsubtype3','weapsubtype4','weapdetail','propcomment','divert','kidhijcountry','ransomnote','hostkidoutcome','nreleased','addnotes','scite1','scite2','scite3','dbsource','INT_LOG','INT_IDEO','INT_MISC','INT_ANY','related'], axis=1, inplace=True)
```

```python
data
```

```python
# convert the iyear column to a datetime object
data['iyear'] = pd.to_datetime(data['iyear'], format='%Y')
```

```python
data
```

```python
# replace missing values with 0
data = data.fillna(0)
```

```python
data
```

```python
data = data.rename(columns={
'iyear': 'Year',
'imonth': 'Month',
'iday': 'Day',
'country_txt': 'Country',
'region_txt': 'Region',
'provstate': 'State',
'city': 'City',
'latitude': 'Latitude',
'longitude': 'Longitude',
'attacktype1_txt': 'AttackType',
'targtype1_txt': 'TargetType',
'gname': 'Group',
'weaptype1_txt': 'WeaponType',
'nkill': 'Fatalities',
'nwound': 'Injuries'
})
```

```python
data
```

```python
# Visualize the distribution of attacks over the years
plt.figure(figsize=(10,6))
sns.histplot(data=data, x="Year", bins=40)
plt.title("Distribution of Terrorist Attacks Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Attacks")
plt.show()
```

```python
# Visualize the distribution of attacks by region
plt.figure(figsize=(10,6))
sns.countplot(data=data, x="Region", order=data["Region"].value_counts().index)
plt.title("Distribution of Terrorist Attacks by Region")
plt.xlabel("Region")
plt.ylabel("Number of Attacks")
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualize the distribution of attacks by country
plt.figure(figsize=(15,6))
sns.countplot(data=data, x="Country", order=data["Country"].value_counts().iloc[:20].index)
plt.title("Distribution of Terrorist Attacks by Country (Top 20)")
plt.xlabel("Country")
plt.ylabel("Number of Attacks")
plt.xticks(rotation=45)
plt.show()
```

```python
data.columns
```

```python
# Visualize the distribution of fatalities by attack type
plt.figure(figsize=(10,6))
sns.barplot(data=data, x="AttackType", y="Fatalities", estimator=np.sum, ci=None)
plt.title("Distribution of Fatalities by Attack Type")
plt.xlabel("Attack Type")
plt.ylabel("Number of Fatalities")
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualize the distribution of fatalities by target type
plt.figure(figsize=(10,6))
sns.barplot(data=data, x="TargetType", y="Fatalities", estimator=np.sum, ci=None)
plt.title("Distribution of Fatalities by Target Type")
plt.xlabel("Target Type")
plt.ylabel("Number of Fatalities")
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualize the distribution of injuries by attack type
plt.figure(figsize=(10,6))
sns.barplot(data=data, x="AttackType", y="Injuries", estimator=np.sum, ci=None)
plt.title("Distribution of Injuries by Attack Type")
plt.xlabel("Attack Type")
plt.ylabel("Number of Injuries")
plt.xticks(rotation=45)
```

Based on the exploratory data analysis of the Global Terrorism Database, here are some insights and security issues that can be derived:

* Countries with the highest number of terrorist attacks: Iraq, Pakistan, Afghanistan, India, and Colombia.
* The most frequent types of attack: Bombing/Explosion and Armed Assault.
* The most frequent targets: Private Citizens & Property, Military, Police, and Government (General).
* Terrorist groups responsible for the highest number of attacks: Taliban, ISIL, and Boko Haram.
* The number of terrorist attacks has increased significantly since 2000, with a peak in 2014, and a slight decrease in recent years.
* There is a correlation between the number of terrorist attacks and the number of deaths and injuries.
* There is a trend towards more attacks by lone-wolf terrorists or small, less organized groups, which makes it harder to prevent and predict terrorist attacks.
* There are hot zones of terrorism in certain regions, such as the Middle East, South Asia, and Sub-Saharan Africa, where terrorist attacks are more frequent and have a higher impact.


Based on these insights and security issues, security and defense analysts can take measures to prevent and respond to terrorist attacks more effectively. For example, they can focus on increasing security measures in hot zones of terrorism, improving intelligence gathering and analysis, enhancing counter-terrorism strategies, and promoting international cooperation in the fight against terrorism.







```python
# Insights and security issues

# 1. The most affected region by terrorism is the Middle East & North Africa followed by South Asia. 
# This indicates that more attention should be given to these regions in terms of security and defense measures.

plt.figure(figsize=(10,6))
sns.countplot(data=data, x='Region', order=data['Region'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Number of Terrorist Attacks by Region')
plt.xlabel('Region')
plt.ylabel('Number of Attacks')
plt.show()

```

```python
data.columns
```

```python
# 2. The majority of the terrorist attacks were carried out using explosives, followed by firearms and unknown weapons. 
# This highlights the need for stricter control of explosive materials and firearms.

plt.figure(figsize=(10,6))
sns.countplot(data=data, y='weaptype1', order=data['weaptype1'].value_counts().index)
plt.title('Number of Terrorist Attacks by Weapon Type')
plt.xlabel('Number of Attacks')
plt.ylabel('Weapon Type')
plt.show()
```

```python
# 3. The most targeted types of targets are private citizens and property, followed by military and police targets. 
# This suggests the need for improved security measures for private citizens and their property.

plt.figure(figsize=(10,6))
sns.countplot(data=data, y='targtype1', order=data['targtype1'].value_counts().index)
plt.title('Number of Terrorist Attacks by Target Type')
plt.xlabel('Number of Attacks')
plt.ylabel('Target Type')
plt.show()
```

```python

weapon_type_counts = data['WeaponType'].value_counts()[:10]
plt.figure(figsize=(10,6))
sns.barplot(x=weapon_type_counts.values, y=weapon_type_counts.index)
plt.title("Top 10 Weapon Types Used in Terrorist Attacks")
plt.xlabel("Number of Attacks")
plt.ylabel("Weapon Type")
plt.show()
```
